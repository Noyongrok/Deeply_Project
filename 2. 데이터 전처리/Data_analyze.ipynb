```python
import pandas as pd
import numpy as np


# 기술 통계를 계산하는 함수
def descriptive_stats(column):
    counts = df[column].value_counts()
    return {
        '평균': counts.mean(),
        '중앙값': np.median(counts),
        '표준편차': counts.std(),
        '최소값': counts.min(),
        '최대값': counts.max()
    }

# 아티스트에 대한 기술 통계 계산
artist_stats = descriptive_stats('artist')

# 앨범에 대한 기술 통계 계산
album_stats = descriptive_stats('album')

# 장르에 대한 기술 통계 계산
genre_stats = descriptive_stats('genre')

# 'release_date'에서 연도 추출 후 2000을 빼서 'release_year' 계산
df['release_year'] = pd.to_numeric(df['release_date'].str.extract(r'(\d{4})')[0], errors='coerce') - 2000
year_stats = descriptive_stats('release_year')

# 통계 출력
print(f"아티스트 통계: {artist_stats}")
print(f"앨범 통계: {album_stats}")
print(f"장르 통계: {genre_stats}")
print(f"발매 연도 통계: {year_stats}")
```

## 시대별 음악 트렌드 

```python
# 연도 컬럼을 생성합니다. 년도 정보가 없는 경우를 대비하여 에러를 'coerce'로 설정합니다.
df['release_year'] = pd.to_numeric(df['release_date'].str.split('-').str[0], errors='coerce')

# 연도별로 데이터를 그룹화합니다.
yearly_data = df.groupby('release_year').agg({'song_name': 'count', 'popularity': 'mean'}).reset_index()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 윈도우에서 사용 가능한 기본 한글 폰트 중 하나를 선택합니다. 예: 'Malgun Gothic'
plt.rc('font', family='Malgun Gothic')

plt.figure(figsize=(14, 7))

# 연도별 곡 수 시각화
plt.subplot(1, 2, 1)
plt.plot(yearly_data['release_year'], yearly_data['song_name'], marker='o')
plt.title('연도별 곡 수')
plt.xlabel('연도')
plt.ylabel('곡 수')

# 연도별 평균 인기도 시각화
plt.subplot(1, 2, 2)
plt.plot(yearly_data['release_year'], yearly_data['popularity'], marker='o', color='r')
plt.title('연도별 평균 인기도')
plt.xlabel('연도')
plt.ylabel('평균 인기도')

plt.tight_layout()
plt.show()

```

## 인기도와 음원 특성의 상관관계

```python
# 인기도와 음원 특성 간의 상관관계를 분석합니다.
features = ['popularity', 'energy', 'danceability', 'valence', 'acousticness', 'instrumentalness', 'liveness', 'speechiness', 'loudness']
correlation_matrix = df[features].corr()

# 상관관계 히트맵을 시각화합니다.
import seaborn as sns

plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('음원 특성 간의 상관관계')
plt.show()
```

- `에너지`가 높으면 `소리의 크기`와의 상관관계 -> 에너지가 높은 곡들이 더 큰소리로 들림 
- `acousticness`특성은 `에너지`와 음의 상관관계  -> 어쿠스틱 곡들이 더 조용하고 차분한 경향이 있음
- 

```python
# 인기도가 높은 곡들을 선택합니다.
popular_songs = df[df['popularity'] > df['popularity'].quantile(0.75)]

# 에너지와 댄스성이 높은 곡들을 시각화합니다.
plt.figure(figsize=(10, 5))
plt.scatter(popular_songs['energy'], popular_songs['danceability'], alpha=0.6, color='blue')
plt.title('에너지와 댄스성이 높은 인기 곡들의 분포')
plt.xlabel('에너지')
plt.ylabel('댄스성')
plt.grid(True)
plt.show()
```

### 장르별 곡 수 분석
