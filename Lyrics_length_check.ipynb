```python
def count_words(sentence):
    words = sentence.split()  # 공백을 기준으로 문자열 분리
    return len(words)  # 분리된 어절의 수 반환
```

```python
def std_word_count(path, file_name, title) :

    data = pd.read_csv(path + file_name)
    data['Word_Count'] = data['lyric'].apply(count_words)
    std_dev = data['Word_Count'].std()

    # 데이터 시각화
    plt.hist(data['Word_Count'], bins=10, alpha=0.7, color='blue', edgecolor='black')
    plt.axvline(data['Word_Count'].mean(), color='red', linestyle='dashed', linewidth=1, label='Mean')
    plt.axvline(data['Word_Count'].mean() + std_dev, color='green', linestyle='dashed', linewidth=1, label='Mean + 1*std_dev')
    plt.axvline(data['Word_Count'].mean() - std_dev, color='green', linestyle='dashed', linewidth=1, label='Mean - 1*std_dev')

    plt.title('{} of Standard Deviation Visualization'.format(title))
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()

    plt.show()
    
    return data
```
