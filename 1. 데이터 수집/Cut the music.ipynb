#파일 묶어서 음원 자르기

```python
import os
from pydub import AudioSegment

def trim_mp3_middle(input_path, output_path, start_time_seconds, end_time_seconds):
    # MP3 파일 로드
    audio = AudioSegment.from_mp3(input_path)

    # 시작 및 끝 시간을 밀리초로 변환
    start_time_ms = start_time_seconds * 1000
    end_time_ms = end_time_seconds * 1000

    # 오디오 길이 체크
    if len(audio) < end_time_ms:
        print(f"The end time exceeds the length of the audio for {input_path}. Skipping...")
        return

    # 오디오 자르기
    trimmed_audio = audio[start_time_ms:end_time_ms]

    # 자른 오디오 저장
    trimmed_audio.export(output_path, format="mp3")

# 폴더 내의 모든 MP3 파일을 탐색
folder_path = "/content/drive/MyDrive/Playlist/1.감성 충전, 힐링 가요"  # 자를 MP3 파일이 있는 폴더 경로
output_folder = "/content/drive/MyDrive/Playlist/1.감성 충전, 힐링 가요/trimmed"  # 자른 MP3 파일을 저장할 폴더 경로

# 출력 폴더가 없으면 생성
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for filename in os.listdir(folder_path):
    if filename.endswith(".mp3"):
        input_path = os.path.join(folder_path, filename)
        output_path = os.path.join(output_folder, f"trimmed_{filename}")

        # 1분 30초 (90초)에서 2분 30초 (150초)까지 자르기
        trim_mp3_middle(input_path, output_path, 30, 90)
```
