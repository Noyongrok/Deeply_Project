# 실행 코드

```python
def search_spotify(songs_info):
    track_ids = []
    failed_requests = []

    for i, (artist, song) in enumerate(songs_info):
        try:
            query = f'{artist} {song}'
            results = sp.search(q=query, type='track', limit=1)
            items = results['tracks']['items']
            if items:
                track_id = items[0]['id']
                track_ids.append(track_id)
            else:
                track_ids.append(None)
        except Exception as e:
            logging.error(f"오류 발생: {e}")
            track_ids.append(None)
            failed_requests.append((i, artist, song))
        time.sleep(3)

    # 재시도 메커니즘
    for i, artist, song in failed_requests:
        try:
            query = f'{artist} {song}'
            results = sp.search(q=query, type='track', limit=1)
            items = results['tracks']['items']
            if items:
                track_id = items[0]['id']
                track_ids[i] = track_id
        except Exception as e:
            logging.error(f"재시도 중 오류 발생: {e}")

    return track_ids

track_ids = search_spotify(songs_info)
```

## 음원 특성데이터 수집(실패한 목록을 리스트에 담아줌 + 한번에 하는 코드)

```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import time

df = pd.read_csv("C:/Users/aischool291/df_for_spotify.csv")
songs_info = df[['artist', 'song']].values.tolist()
```

```python
def search_and_get_features(songs_info, batch_size=100):
    total_tracks = len(songs_info)
    features_list = []
    failed_searches = []  # 실패한 검색을 기록할 리스트
    failed_tracks = []    # 특성 추출에 실패한 트랙들을 기록할 리스트

    for start in range(0, total_tracks, batch_size):
        end = start + batch_size
        print(f"처리 시작: 배치 {start} - {end}")

        batch = songs_info[start:end]
        for artist, song in batch:
            try:
                # 트랙 ID 검색
                query = f'{artist} {song}'
                results = sp.search(q=query, type='track', limit=1)
                track_id = results['tracks']['items'][0]['id'] if results['tracks']['items'] else None

                if track_id:
                    # 특성 추출
                    audio_features = sp.audio_features(track_id)[0]
                    track_info = sp.track(track_id)

                    if audio_features and track_info:
                        features = {
                            'artist': artist,
                            'song': song,
                            'track_id': track_id,
                            'energy': audio_features['energy'],
                            'danceability': audio_features['danceability'],
                            'valence': audio_features['valence'],
                            'acousticness': audio_features['acousticness'],
                            'instrumentalness': audio_features['instrumentalness'],
                            'speechiness': audio_features['speechiness'],
                            'loudness': audio_features['loudness'],
                        }
                        features_list.append(features)
                    else:
                        failed_tracks.append((artist, song))
                else:
                    failed_searches.append((artist, song))
            except Exception as e:
                logging.error(f"배치 {start} - {end}에서 오류 발생: {e}")
                failed_searches.append((artist, song))
            time.sleep(2.5)

        print(f"처리 완료: 배치 {start} - {end}")

    return pd.DataFrame(features_list), failed_searches, failed_tracks

# 함수 호출
df_features, failed_searches, failed_tracks = search_and_get_features(songs_info)

# 결과 출력
print("실패한 검색 목록:", failed_searches)
print("실패한 트랙들:", failed_tracks)
print(df_features)  # 추출된 특성 데이터 프레임 출력
```
